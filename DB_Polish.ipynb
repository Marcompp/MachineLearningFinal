```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
print("Hello World")
```

```python
df_yugi = pd.read_csv('dbs/yugi_comp.csv')
df_yugi = df_yugi.drop(['name'], axis=1)
df_yugi = df_yugi[df_yugi['text'].notna()]
df_yugi.head()
```

```python
def getTribe(row):
    #if row["card_type"] == "monster":
    if type(row["type"]) is float:
        return math.nan
    return row["type"].split("/")[0].strip()
    #else:
    #    return math.nan
```

```python
def getEffect(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")[-1].strip()
        if typ == "Effect":
            return typ
        else:
            return "Normal"
    else:
        return math.nan
```

```python
extradeckmons = ['Xyz', 'Fusion', 'Synchro', 'Ritual', 'Toon', 'Link', 'Union', 'Gemini', 'Skill']
def getMonsType(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        if len(typ) >= 3 and typ[1].strip() in extradeckmons:
            return typ[1].strip()
    else: 
        return row["type"]
    return math.nan
```

```python
df_yugi['tribe'] = df_yugi.apply(lambda row: getTribe(row), axis=1)
df_yugi.head()
```

```python
df_yugi['effect?'] = df_yugi.apply(lambda row: getEffect(row), axis=1)
df_yugi.head()
```

```python
df_yugi['monsType'] = df_yugi.apply(lambda row: getMonsType(row), axis=1)
df_yugi.head()
```

```python
extradeckmons = ['Xyz', 'Fusion', 'Synchro', 'Ritual', 'Toon', 'Link', 'Union', 'Gemini', 'Skill']

othertype  = ['Spirit', 'Continuous Spell' 'Field Spell','Flip','Continuous Trap']

def getMonsType2(row):
    monst = ""
    if type(row["type"]) is float:
        return math.nan
    #print(row["type"])
    for word in row["type"].split("/"):
            if word.strip() in othertype:
                monst += word.strip() 
    if monst == "":
        return math.nan
    else:
        return monst
```

```python
def getMonsType3(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        for ty in typ:
            typp = ty.strip()
            if typp == "Tuner":
                return "Tuner"
        
    return math.nan
```

```python

def getMonsType4(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        for ty in typ:
            typp = ty.strip()
            if typp == "Pendulum":
                return "Pendulum"
        
    return math.nan
```

```python
df_yugi['monsType2'] = df_yugi.apply(lambda row: getMonsType2(row), axis=1)
df_yugi['Tuner?'] = df_yugi.apply(lambda row: getMonsType3(row), axis=1)
df_yugi['Pendulum?'] = df_yugi.apply(lambda row: getMonsType4(row), axis=1)
df_yugi.head()
```

```python

import tensorflow_hub as hub

model = hub.KerasLayer("https://tfhub.dev/google/nnlm-en-dim128/2")
```

```python
embeddings=model(df_yugi["text"])
```

```python
embeddings.shape
```

```python
from sklearn.decomposition import PCA

pca_yugi = PCA(n_components=5)
principalComponents_yugi = pca_yugi.fit_transform(embeddings)
principalComponents_yugi
```

```python
df = pd.DataFrame(principalComponents_yugi, columns=['pca1', 'pca2','pca3','pca4','pca5'])
df
```

```python
df1 = df_yugi.copy()
df1 = df1.join(df)
df1
```

```python
df1.drop(columns=['Unnamed: 0', 'type', 'isGood'], inplace=True)
```

```python
mapper_dict = { 
                'text': 'Description', 
                'card_type': 'CardType',
                'family': 'Attribute',
                'atk': 'ATK',
                'def': 'DEF', 
                'level': 'LVL', 
                'property': 'Property', 
                'tribe': 'MonsterType',
                'effect?': 'isEffect',
                'monsType': 'Ability1',
                'monsType2': 'Ability2',
                'Tuner?': 'isTuner',
                'Pendulum?': 'isPendulum',
                'pca1': 'PCA1',
                'pca2': 'PCA2',
                'pca3': 'PCA3',
                'pca4': 'PCA4',
                'pca5': 'PCA5'
              }

df1.rename(mapper=mapper_dict, axis=1, inplace=True)
```

```python
df1.to_csv("dbs/yugi_db_polished.csv")  
```
