```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
print("Hello World")
```

```python
df_yugi = pd.read_csv('dbs/yugi_comp.csv', index_col=0)
df_yugi.drop(['name'], inplace=True, axis=1)
df_yugi = df_yugi[df_yugi['text'].notna()]
df_yugi.head()
```

```python
def getTribe(row):
    #if row["card_type"] == "monster":
    if type(row["type"]) is float:
        return math.nan
    return row["type"].split("/")[0].strip()
    #else:
    #    return math.nan
```

```python
def getEffect(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")[-1].strip()
        if typ == "Effect":
            return typ
        else:
            return "Normal"
    else:
        return math.nan
```

```python
extradeckmons = ['Xyz', 'Fusion', 'Synchro', 'Ritual', 'Toon', 'Link', 'Union', 'Gemini', 'Skill']
def getMonsType(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        if len(typ) >= 3 and typ[1].strip() in extradeckmons:
            return typ[1].strip()
    else: 
        return row["type"]
    return math.nan
```

```python
df_yugi['tribe'] = df_yugi.apply(lambda row: getTribe(row), axis=1)
df_yugi.head()
```

```python
df_yugi['effect?'] = df_yugi.apply(lambda row: getEffect(row), axis=1)
df_yugi.head()
```

```python
df_yugi['monsType'] = df_yugi.apply(lambda row: getMonsType(row), axis=1)
df_yugi.head()
```

```python
extradeckmons = set(df_yugi["monsType"].unique())
extradeckmons.remove(math.nan)
extradeckmons
```

```python
# extradeckmons = ['Xyz', 'Fusion', 'Synchro', 'Ritual', 'Toon', 'Link', 'Union', 'Gemini', 'Skill']

othertype  = ['Spirit', 'Continuous Spell' 'Field Spell','Flip','Continuous Trap']

def getMonsType2(row):
    monst = ""
    if type(row["type"]) is float:
        return math.nan
    #print(row["type"])
    for word in row["type"].split("/"):
            if word.strip() in othertype:
                monst += word.strip() 
    if monst == "":
        return math.nan
    else:
        return monst
```

```python
def getMonsType3(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        for ty in typ:
            typp = ty.strip()
            if typp == "Tuner":
                return "Tuner"
        
    return math.nan
```

```python

def getMonsType4(row):
    if row["card_type"] == "monster":
        typ =  row["type"].split("/")
        for ty in typ:
            typp = ty.strip()
            if typp == "Pendulum":
                return "Pendulum"
        
    return math.nan
```

```python
df_yugi['monsType2'] = df_yugi.apply(lambda row: getMonsType2(row), axis=1)
df_yugi['Tuner?'] = df_yugi.apply(lambda row: getMonsType3(row), axis=1)
df_yugi['Pendulum?'] = df_yugi.apply(lambda row: getMonsType4(row), axis=1)
df_yugi.head()
```
